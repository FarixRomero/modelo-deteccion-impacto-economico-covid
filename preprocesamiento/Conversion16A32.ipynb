```python
from osgeo import gdal, gdalconst

# Ruta de la imagen de entrada (16 bits)
input_image_path = '/home/farix/python/dataset/imagenesTotalesTiff/IMG_PER1_20170121_block_1_1_block_0_0.tif'

# Ruta de la imagen de salida (32 bits) sin la cuarta banda (NIR)
output_image_path = 'imagen_convertida.tif'

# Abrir la imagen de entrada
input_ds = gdal.Open(input_image_path, gdalconst.GA_ReadOnly)

# Crear una nueva imagen con solo las tres primeras bandas (rojo, verde y azul)
output_ds = gdal.GetDriverByName('GTiff').Create(output_image_path, input_ds.RasterXSize, input_ds.RasterYSize, 3, gdalconst.GDT_Float32)

# Copiar los datos de las tres primeras bandas a la nueva imagen
for band_num in range(1, 4):
    input_band = input_ds.GetRasterBand(band_num)
    output_band = output_ds.GetRasterBand(band_num)
    output_band.WriteArray(input_band.ReadAsArray())

# Cerrar los conjuntos de datos
input_ds = None
output_ds = None

print("Conversión completada.")
```

```python
import os
from osgeo import gdal, gdalconst

# Carpeta de entrada con las imágenes TIFF de 16 bits
input_folder = '/home/farix/python/dataset/imagenesTotalesTiff'

# Carpeta de salida para las imágenes convertidas
output_folder = '/home/farix/python/dataset/imagenesTotalesTiff32'

# Lista todos los archivos en la carpeta de entrada
input_files = [f for f in os.listdir(input_folder) if f.endswith('.tif')]

# Recorre cada archivo en la carpeta de entrada
for input_file in input_files:
    input_image_path = os.path.join(input_folder, input_file)
    
    # Genera la ruta de salida
    output_file = input_file.replace('.tif', '.tif')
    output_image_path = os.path.join(output_folder, output_file)

    # Abrir la imagen de entrada
    input_ds = gdal.Open(input_image_path, gdalconst.GA_ReadOnly)

    # Crear una nueva imagen con solo las tres primeras bandas (rojo, verde y azul)
    output_ds = gdal.GetDriverByName('GTiff').Create(output_image_path, input_ds.RasterXSize, input_ds.RasterYSize, 3, gdalconst.GDT_Float32)

    # Copiar los datos de las tres primeras bandas a la nueva imagen
    for band_num in range(1, 4):
        input_band = input_ds.GetRasterBand(band_num)
        output_band = output_ds.GetRasterBand(band_num)
        output_band.WriteArray(input_band.ReadAsArray())

    # Copiar la información de georreferenciación
    output_ds.SetGeoTransform(input_ds.GetGeoTransform())
    output_ds.SetProjection(input_ds.GetProjection())

    # Cerrar los conjuntos de datos
    input_ds = None
    output_ds = None

    print(f"Conversión completada para {input_file}")
```
