```python

import numpy as np

def pansharpening_quarter(pan_path, rgb_path, output_path):
    # Abrir la imagen pancromática (escala de grises)
    pan_dataset = gdal.Open(pan_path, gdal.GA_ReadOnly)
    
    # Obtener información de la imagen pancromática
    pan_geotransform = pan_dataset.GetGeoTransform()
    pan_projection = pan_dataset.GetProjection()
    
    # Obtener el tamaño de la imagen pancromática
    pan_width = pan_dataset.RasterXSize
    pan_height = pan_dataset.RasterYSize
    
    # Cerrar el dataset de la imagen pancromática
    pan_dataset = None
    
    # Abrir la imagen RGB
    rgb_dataset = gdal.Open(rgb_path, gdal.GA_ReadOnly)
    
    # Obtener información de la imagen RGB
    rgb_geotransform = rgb_dataset.GetGeoTransform()
    rgb_projection = rgb_dataset.GetProjection()
    
    # Obtener el tamaño de la imagen RGB
    rgb_width = rgb_dataset.RasterXSize
    rgb_height = rgb_dataset.RasterYSize
    
    # Cerrar el dataset de la imagen RGB
    rgb_dataset = None
    
    # Definir el tamaño de un cuarto de ambas imágenes
    quarter_width = pan_width // 2
    quarter_height = pan_height // 2
    
    # Leer solo un cuarto de la imagen pancromática y almacenarla en la matriz
    pan_dataset = gdal.Open(pan_path, gdal.GA_ReadOnly)
    pan_array = pan_dataset.ReadAsArray(0, 0, quarter_width, quarter_height)
    pan_dataset = None
    
    # Leer solo un cuarto de la imagen RGB y almacenarla en la matriz
    rgb_dataset = gdal.Open(rgb_path, gdal.GA_ReadOnly)
    rgb_array = np.zeros((3, quarter_height, quarter_width), dtype=np.float32)
    for i in range(3):
        band = rgb_dataset.GetRasterBand(i + 1)
        rgb_array[i, :, :] = band.ReadAsArray(0, 0, quarter_width, quarter_height)
    rgb_dataset = None
    
    # Realizar el proceso de pansharpening en el cuarto de ambas imágenes
    pan_weight = 0.5
    rgb_weight = 0.5
    pansharpened_quarter = (pan_weight * pan_array) + (rgb_weight * np.mean(rgb_array, axis=0))
    
    # Crear una nueva imagen RGB con la resolución de la imagen pancromática
    driver = gdal.GetDriverByName('GTiff')
    output_dataset = driver.Create(output_path, quarter_width, quarter_height, 3, gdal.GDT_Float32)
    
    # Escribir los datos en la nueva imagen RGB
    for i in range(3):
        band = output_dataset.GetRasterBand(i + 1)
        band.WriteArray(pansharpened_quarter)
    
    # Establecer información geoespacial y proyección
    output_dataset.SetGeoTransform(pan_geotransform)
    output_dataset.SetProjection(pan_projection)
    
    # Cerrar el dataset de la nueva imagen RGB
    output_dataset = None
    
    print("Pansharpening completado en un cuarto de ambas imágenes y se ha generado una imagen RGB con la resolución de la imagen pancromática.")
    print("La imagen resultante se ha guardado en:", output_path)

# Ejemplo de uso
pan_path = 'imagenes/CO_2305200113402/VOL_PER1_ORT_001_000054/IMG_PER1_ORT_P_000054/IMG_PER1_20230129153659_ORT_P_000054.TIF'
rgb_path = 'imagenes/CO_2305200113402/VOL_PER1_ORT_001_000054/IMG_PER1_ORT_MS_000054/IMG_PER1_20230129153659_ORT_MS_000054_RESCALADO.TIF'  # Ruta de la imagen RGB reescalada
output_path = 'imagenes/Rescalado.tif'
pansharpening_quarter(pan_path, rgb_path, output_path)
```
