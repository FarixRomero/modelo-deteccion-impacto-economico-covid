```python
# Imagenes tif de Perusat (4 bandas RGB NIR) a PNG (3 bandas RGB)

import os
import errno
import rasterio
from pathlib import Path

# Listo los archivos con os.listdir
# La ruta son los archivos tif
folder_path_tif = '/home/farix/python/imagen/tiff/'

# cambio la ruta para guardar las imagenes png
# cambio tiff por png, para tener una carpeta con mismo nombres de las imagenes tif
folder_path_png = folder_path_tif.replace('tiff', 'png')

# imprimo la ruta del folder png
print(folder_path_png)

# creo folder para guardar las imagenes png
try:
    os.mkdir(folder_path_png)
except OSError as e:
    if e.errno != errno.EEXIST:
        raise

# Carpeta de salida
#folder_path_png = '/home/ferdinand/Documents/dataset/perusat_Sentinel_ort/split_perusat/png/' + file_name + '/'
#print(folder_path_png)

#folder_path_png = '/home/ferdinand/Documents/dataset/perusat_Sentinel_ort/split_perusat/png/IMG_PER1_20190521155848_ORT_MS_000401/'

for entry in os.listdir(folder_path_tif):
    if os.path.isfile(os.path.join(folder_path_tif, entry)):
        print(entry)
        # Para GDAL, tomo los max y min de cada banda en la imagen TIF
        # Con rasterio
        RGB = os.path.join(folder_path_tif, entry)
        imagen_tif = rasterio.open(RGB)

        # lectura del min scale  banda 1
        arrayMin1 = imagen_tif.read(1)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda1Min = arrayMin1.min()
        # lectura del max scale  banda 1
        arrayMax1 = imagen_tif.read(1)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda1Max = arrayMax1.max()

        # lectura del min scale  banda 2
        arrayMin2 = imagen_tif.read(2)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda2Min = arrayMin2.min()

        # lectura del max scale  banda 2
        arrayMax2 = imagen_tif.read(2)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda2Max = arrayMax2.max()

        # lectura del min scale  banda 3
        arrayMin3 = imagen_tif.read(3)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda3Min = arrayMin3.min()

        # lectura del max scale  banda 3
        arrayMax3 = imagen_tif.read(3)
        # banda1Min = band.min
        # Mostrar stado de cada banda
        banda3Max = arrayMax3.max()

        # Conversion a PNG
        
        # Creo carpeta para los archivos de salida con el nombre de la imagen de entrada
        #file_name = Path(entry).stem
        #print(file_name)

        #try:
        #    os.mkdir(folder_path_png)
        #except OSError as e:
        #    if e.errno != errno.EEXIST:
        #        raise
        
        # La carpeta de salida de los archivos png
        # archivo de salida, el nombre es el mismo que el archivo de entrada con extension PNG
        
        without_ext = os.path.splitext(entry)[0]
        file_PNG = folder_path_png + without_ext + '.png'
        print(file_PNG)

        # Genero una linea para ser ejecutada en consola para la conversion de tif a PNG
        # Para enviar a bash el comando y realize la conversion de tif a png con gdal
        # Se debe considerar el orden de las bandas de perusat
        # blue = banda1
        # green = Banda2
        # red = banda3
        # nir = banda4

        # Una imagen png
        # Red = banda1
        # Green = banda2
        # Blue = banda3

        # Se invierte los colores al tomar la imagen TIF : -b 1 -b 2 -b 3

        #to_png = ('gdal_translate -ot Byte -of PNG -b 1 -b 2 -b 3 -scale_1 ' + str(banda1Min) + ' ' + str(
        #    banda1Max) + ' -scale_2 ' + str(banda2Min) + ' ' + str(banda2Max) + ' -scale_3 ' + str(
        #    banda3Min) + ' ' + str(banda3Max) + ' -a_nodata none ' + RGB + ' ' + file_PNG + ' -co PHOTOMETRIC=RGB')

        to_png = ('gdal_translate -ot Byte -of PNG -b 1 -b 2 -b 3 -scale_1 ' + str(banda1Min) + ' ' + str(
            banda1Max) + ' -scale_2 ' + str(banda2Min) + ' ' + str(banda2Max) + ' -scale_3 ' + str(
            banda3Min) + ' ' + str(banda3Max) + ' -a_nodata none ' + RGB + ' ' + file_PNG)
        print(to_png)
        # Ejecuto en un bash de linux para la conversion
        os.system(to_png)
        
#Remuevo los archivos xml
delete_xml = "rm " + folder_path_png + "*.aux.xml"
print(delete_xml)
os.system(delete_xml)        
        
```
