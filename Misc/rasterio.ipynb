```python
from PIL import Image
import os
import numpy as np

def tiff_to_jpg(input_path, output_path):
    # Abre la imagen TIFF
    img = Image.open(input_path)

    # Divide la imagen en bandas
    bands = img.split()

    # Define los rangos de valores mínimos y máximos para cada banda de Perusat
    min_max_ranges_perusat = [(357, 681), (436, 617), (560, 736)]

    normalized_band_arrays_rgb = []

    for i in range(3):
        band = np.array(bands[i], dtype=np.uint16)
        
        if "perusat" in input_path.lower():
            min_val, max_val = min_max_ranges_perusat[i]
        else:
            # Si es una imagen JPG, se asume que ya está en el formato adecuado (0-255)
            min_val, max_val = 0, 255
        
        # Normaliza los datos de la banda al rango de 0 a 255
        normalized_band = ((band - min_val) / (max_val - min_val) * 255).astype(np.uint8)
        normalized_band_arrays_rgb.append(normalized_band)

    # Reordena las bandas para que coincidan con el orden de una imagen JPG (RGB)
    rgb_image = Image.merge("RGB", (Image.fromarray(normalized_band_arrays_rgb[2]),
                                    Image.fromarray(normalized_band_arrays_rgb[1]),
                                    Image.fromarray(normalized_band_arrays_rgb[0])))

    # Guarda la imagen RGB en formato JPG
    rgb_image.save(output_path, "JPEG", quality=95)

if __name__ == "__main__":
    input_tiff_path = "output_blocks/block_1_1.tif"
    output_jpg_path = "bloque.jpg"

    tiff_to_jpg(input_tiff_path, output_jpg_path)
```
